## Preguntas, análisis y respuestas

1)	¿Cómo se distribuye la cantidad de hijos nacidos en la población de muestra?(s4)
Se filtra la variable s4, para evitar que se muestre el valor 99, pues este dato no hace referencia a una cantidad de hijos nacidos.

```python
casen_nacidos=casen[casen['s4']<80]
print("Resumen de 5 números de la variable s4")
print("-"*40)
print(casen_nacidos['s4'].describe())
plt.figure(figsize=(4,4))
sns.boxplot(x=casen_nacidos['s4'])
plt.title('Diagrama de caja de "La cantidad de hijos nacidos vivios"',y=1.01)
plt.xlabel('n° de hijos')
plt.show()
```

Se observa que el promedio es de 1.777 hijos, cercano a 2 hijos, además, que la mediana se encuentra en 2 hijos. A su vez, se observa la existencia de datos atípicos, incluso hay personas que han tenido más de 20 hijos.

**Respuestas:** Se tiene que la distribución es más densa entre 0 y 3 hijos, con una desviación estandar de 1.9 y un rango de 22 hijos.

2)	¿Existe diferencia significa en la distribución de las personas que han tenido otro trabajo, o no, aparte de su trabajo principal (o26), según el número de hijos que han tenido (s4)?


```python
plt.figure(figsize=(10,4))
count_plot = sns.countplot(x = casen.s4, hue = casen.o26)
plt.title('Gráfico de barras para "La cantidad de hijos nacidos vivos" por "Haber tenido un trabajo aparte del principal"',y=1.01)
plt.ylabel('n° de personas')
plt.xlabel('n° de hijos')
plt.show()
```

Del gráfico se interpreta que hay más personas que no han tenido trabajo extra, la mayor cantidad de respuestas se dan en personas que han tenido entre 0 y 3 hijos. 

**Respuesta:** No se observa una diferencia en la tendencia de la distribución, pues sucede que tanto las personas que han tenido un trabajo como las que no, aumentan o disminuyen de la misma forma con la cantidad de hijos.

3)	¿Cómo se distribuye la edad al tener el primer hijo? (Variable a usar s5 filtrada, pues 99 no es un dato que entregue una edad)

```python
casen_edad_n=casen[casen['s5']<90]

print("Resumen de 5 números de la variable s5")
print("-"*40)
print(casen_edad_n['s5'].describe())

plt.figure(figsize=(8,4))
plot = plt.hist(casen_edad_n.s5, bins = 15,density=False,edgecolor = 'black') # bins corresponde al num de segmentos
plt.title('Histograma de "La Edad a la que tuvo a su primer hijo"',y=1.01)
plt.ylabel('n° de personas por cada segmento')
plt.xlabel('Edad')
plt.show()
```

Del histrograma se **interpresta** que los datos estan concentrados en la edad de 20 años, sobre los 40 son muy pocas las personas que tienen su primer hijo.

**Respuesta** Se observa que el promedio está cercano a la mediana, 23.6 y 23 años respectivamente, y que la mayor cantidad de datos se encuentran entre 19 y 27 años aproximadamente.

4)	¿Cuál es el número de personas que tenían 20 años o menos al tener su primer hijo vivo (s5) y si se ha realizado mamografía en los últimos 3 años (s10)?

```python
casen_edad_n=casen[casen['s5']<=20]
casen_mamog_si=casen[casen['s10']<=3]
print('*' * 20, "Tabla de frecuencia de las variables s10 vs s5", '*' * 20)
tabla1=pd.crosstab(casen_edad_n['s5'],casen_mamog_si['s10'],margins=True,margins_name='Suma')
print(tabla1)
```

**Respuesta:** Mediante el filtrado y el uso de una tabla de frecuencias cruzadas se llega a que el total es de 12559 personas.

5)	¿Cómo se distribuye el área de las viviendas de las personas encuestadas (v12) comparando si terminaron Educación Media Cientifico-Humanista vs si terminaron su Educación Profesional Completa  (e6a)?

```python
fig = plt.figure(figsize=(4, 8))

fig.subplots_adjust(hspace=0.4)
                   
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

casen_media=casen[casen['e6a']==9]
casen_pro=casen[casen['e6a']==12]

sns.countplot(x =casen_media['v12'], hue =casen_media['e6a'],ax=ax1)
ax1.set_title('Gráfico de barras para "Categorías de áreas de las vivendas" y "Educación Media C-H" ',y=1)
ax1.set_ylabel('n° de personas')
ax1.set_xlabel('Categorías de áreas de las vivienda')

sns.countplot(x =casen_pro['v12'], hue =casen_pro['e6a'],ax=ax2)
ax2.set_title('Gráfico de barras para "Categorías de áreas de las vivendas" con "Educación Profesional"',y=1)
ax2.set_ylabel('n° de personas')
ax2.set_xlabel('Categorías de áreas de las vivienda')

plt.show()
```

Se **interpreta** de los graficos que hay mayor cantidad de respuestas de personas que han terminado la educación media que las que han terminado la educación profesional, en propociones se observa similitud, sin embargo, se tiene que las personas profesionales aumentan en proporción el tamaño de su vivienda entre 61-100 metros cuadrados. 

**Respuesta:** La distribución es muy similar, exceptuando el aumento proporcional de 61-100 metros cuadrados por parte de los y las profesionales.

11)	¿Las personas se van de la comuna en que nacieron? (r1b) 

```python
personas = casen['r1b'].value_counts()
desfase = (0.1,0.1,0.25,.3)

plt.figure(figsize=(4,4)) #plt. hace referencia a la librería matplotlib
plt.pie(personas, labels=["En esta comuna","En otra comuna de Chile","En otro país", "No sabe"], autopct='%0.1f%%',explode=desfase)
plt.title('Distribución de "La comuna en donde vivía su madre cuando nació usted"',y=1.1)
plt.axis('equal')
plt.show()
```

Se **interpreta** mediante el gráfico que el 59,5% de las personas encuestadas se mantiene en la comuna donde nacieron, el 36.5% se va de la comuna y un porcentaje mínimo es de otro país o no sabe.

**Respuestas:** No, la mayoria de las personas encuestadas permanecen en la comuna donde nacieron.
